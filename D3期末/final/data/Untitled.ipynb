```python
import json
with open('world-50m.simplified.json') as f:
    geo = json.load(f)
with open('countries.json') as fc:
    countries = json.load(fc)
    
print(geo)
```

```python
cca = {}
for c in countries:
    cca[c['cca3']] = c['cca2']
```

```python
for g in geo['objects']['countries']['geometries']:
    if g['id'] in cca:
        g['id'] = cca[g['id']]

print(geo)
```

```python
with open('countries.geo.new1.json', 'w') as ng:
    json.dump(geo,ng)
```
